```python
#
# FIND MIN ELEMENT
#

n = 10
l = [random.randint(-10,10) for i in range(n)]

min_el = l[0]
for i in range(1, len(l)):
    if(min_el > l[i]):
        min_el = l[i]

print(l)
print(min(l))
print(min_el)
```
